```python
# Advance Array Manipulation
# Stacking Array
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

```python
# Splitting array
c = np.arange(16).reshape(4,4)
c
```

```python
np.split(c,4)
```

```python
np.vsplit(c,2)
```

```python
np.hsplit(c,4)
```

```python
a = np.hsplit(c,4)
for i in a:
    print(i)
```
